```
# 1. SETUP: Install all required libraries
!pip install pytubefix gradio nodejs-wheel-binaries
!apt-get install -y ffmpeg

import gradio as gr
import os
from pytubefix import YouTube
from pytubefix.cli import on_progress

# Function to handle the download logic
def colab_download(url, client_type):
    if not url:
        return "Please enter a URL", None

    try:
        # Initialize YouTube with specific client to bypass bot detection
        # WEB usually works, but ANDROID is a great fallback for Colab IPs
        yt = YouTube(url, client=client_type)

        # Get the highest resolution progressive stream (Video + Audio in one file)
        # This replaces the hardcoded index[1] for better reliability
        stream = yt.streams.get_highest_resolution()

        if not stream:
            return "No suitable stream found.", None

        # Download the file
        out_file = stream.download()

        # Return success message and the file for the browser to download
        return f"‚úÖ Success: {yt.title}", out_file

    except Exception as e:
        error_msg = str(e)
        if "403" in error_msg:
            return "‚ùå Error 403: YouTube blocked this request. Try switching 'Client Type' to ANDROID.", None
        return f"‚ùå Error: {error_msg}", None

# 2. UI DESIGN: Create the Gradio Frontend
with gr.Blocks(theme=gr.themes.Default(primary_hue="red", secondary_hue="gray")) as demo:
    gr.Markdown("# üì∫ YouTube Downloader for Google Colab")
    gr.Markdown("Paste your link below. If 'WEB' fails, switch to 'ANDROID'.")

    with gr.Row():
        url_input = gr.Textbox(
            label="YouTube URL",
            placeholder="https://www.youtube.com/watch?v=..."
        )
        client_choice = gr.Radio(
            ["WEB", "ANDROID", "MWEB"],
            label="Client Type (Bypass Method)",
            value="WEB"
        )

    download_btn = gr.Button("üöÄ Download Now", variant="primary")

    with gr.Column():
        status_output = gr.Textbox(label="Status Report")
        file_output = gr.File(label="Download Your File Here")

    # Link the button to the function
    download_btn.click(
        fn=colab_download,
        inputs=[url_input, client_choice],
        outputs=[status_output, file_output]
    )

# 3. LAUNCH: This generates a local and public link
demo.launch(share=True, debug=True)
```

```
# 1. SETUP: Install all required libraries
!pip install pytubefix gradio nodejs-wheel-binaries
!apt-get install -y ffmpeg

import gradio as gr
import os
import subprocess
from pytubefix import YouTube

def download_engine(url, client_type, quality_type):
    if not url:
        return "Please enter a URL", None

    try:
        yt = YouTube(url, client=client_type)
        video_title = "".join(x for x in yt.title if x.isalnum() or x in "._- ") # Sanitize filename

        if quality_type == "Fastest (720p max)":
            # Progressive streams have video and audio combined but cap at 720p
            stream = yt.streams.get_highest_resolution()
            final_file = stream.download(filename=f"{video_title}_fast.mp4")
            return f"‚úÖ Success: {yt.title} (720p)", final_file

        else:
            # HIGHEST QUALITY LOGIC (1080p, 4K, etc.)
            # 1. Get highest video-only stream
            v_stream = yt.streams.filter(adaptive=True, file_extension='mp4', only_video=True).order_by('resolution').desc().first()
            # 2. Get highest audio-only stream
            a_stream = yt.streams.filter(adaptive=True, file_extension='mp4', only_audio=True).order_by('abr').desc().first()

            if not v_stream or not a_stream:
                return "Could not find high-quality streams.", None

            # 3. Download both temporarily
            v_path = v_stream.download(filename="temp_video.mp4")
            a_path = a_stream.download(filename="temp_audio.mp4")

            # 4. Use FFmpeg to merge them
            output_filename = f"{video_title}_HQ.mp4"
            # cmd: ffmpeg -i video.mp4 -i audio.mp4 -c copy output.mp4
            merge_command = f'ffmpeg -y -i "{v_path}" -i "{a_path}" -c copy "{output_filename}"'
            subprocess.run(merge_command, shell=True)

            # 5. Clean up temp files
            if os.path.exists("temp_video.mp4"): os.remove("temp_video.mp4")
            if os.path.exists("temp_audio.mp4"): os.remove("temp_audio.mp4")

            return f"‚úÖ Success: {yt.title} ({v_stream.resolution})", output_filename

    except Exception as e:
        return f"‚ùå Error: {str(e)}", None

# 2. UI DESIGN
with gr.Blocks(theme=gr.themes.Soft(primary_hue="orange")) as demo:
    gr.Markdown("# üé• Ultimate Colab YouTube Downloader")
    gr.Markdown("Choose **Highest Quality** for 1080p/4K (uses FFmpeg to merge video + audio).")

    with gr.Row():
        url_input = gr.Textbox(label="YouTube URL", placeholder="Paste link...")
        client_choice = gr.Radio(["WEB", "ANDROID"], label="Bypass Method", value="WEB")

    quality_choice = gr.Radio(
        ["Fastest (720p max)", "Highest Quality (1080p/4K - Merged)"],
        label="Quality Mode",
        value="Highest Quality (1080p/4K - Merged)"
    )

    download_btn = gr.Button("üöÄ Download", variant="primary")

    with gr.Column():
        status_output = gr.Textbox(label="Status")
        file_output = gr.File(label="Download Result")

    download_btn.click(
        fn=download_engine,
        inputs=[url_input, client_choice, quality_choice],
        outputs=[status_output, file_output]
    )

# 3. LAUNCH
demo.launch(share=True, debug=True)
```

```

import gradio as gr
import os
import subprocess
from pytubefix import YouTube

def download_engine(url, quality_type):
    if not url:
        return "Please enter a URL", None

    try:
        # use_oauth=True triggers the login flow in the Colab terminal/logs
        # allow_oauth_cache=True saves your login so you don't do it every time
        yt = YouTube(url, use_oauth=True, allow_oauth_cache=True)

        video_title = "".join(x for x in yt.title if x.isalnum() or x in "._- ")

        if quality_type == "Fastest (720p max)":
            stream = yt.streams.get_highest_resolution()
            final_file = stream.download(filename=f"{video_title}_720p.mp4")
            return f"‚úÖ Success: {yt.title}", final_file

        else:
            # HIGHEST QUALITY (Merging 1080p/4K Video + Audio)
            v_stream = yt.streams.filter(adaptive=True, file_extension='mp4', only_video=True).order_by('resolution').desc().first()
            a_stream = yt.streams.filter(adaptive=True, file_extension='mp4', only_audio=True).order_by('abr').desc().first()

            if not v_stream or not a_stream:
                return "Could not find adaptive streams. Try Fastest mode.", None

            v_path = v_stream.download(filename="temp_v.mp4")
            a_path = a_stream.download(filename="temp_a.mp4")

            output_filename = f"{video_title}_Highest.mp4"
            merge_command = f'ffmpeg -y -i "{v_path}" -i "{a_path}" -c copy "{output_filename}"'
            subprocess.run(merge_command, shell=True)

            if os.path.exists("temp_v.mp4"): os.remove("temp_v.mp4")
            if os.path.exists("temp_a.mp4"): os.remove("temp_a.mp4")

            return f"‚úÖ Success: {yt.title} ({v_stream.resolution})", output_filename

    except Exception as e:
        return f"‚ùå Error: {str(e)} \n\nCheck the Colab terminal below the cell to see if you need to authorize Google.", None

# 2. UI DESIGN
with gr.Blocks(theme=gr.themes.Default(primary_hue="blue")) as demo:
    gr.Markdown("# üõ°Ô∏è YouTube Downloader (Bypass 403)")
    gr.Markdown("### ‚ö†Ô∏è IMPORTANT: Look at the Colab Output below the cell after clicking Download.")
    gr.Markdown("If it's your first time, you will see a link and a code. Visit the link, enter the code, and login to authorize.")

    url_input = gr.Textbox(label="YouTube URL")
    quality_choice = gr.Radio(
        ["Fastest (720p max)", "Highest Quality (Merged)"],
        label="Quality Choice",
        value="Highest Quality (Merged)"
    )

    download_btn = gr.Button("üöÄ Start Download", variant="primary")

    status_output = gr.Textbox(label="Status")
    file_output = gr.File(label="Download File")

    download_btn.click(
        fn=download_engine,
        inputs=[url_input, quality_choice],
        outputs=[status_output, file_output]
    )

demo.launch(share=True, debug=True)
```
